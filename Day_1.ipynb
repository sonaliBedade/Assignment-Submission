```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[0,2,4,6,8]

#Resize the graph
plt.figure(figsize=(10,8))

#Plot a line graph
plt.plot(x,y,label="line",marker='o',color='g',markeredgecolor='r',markersize=10)

#Add a title, legend, label to your graph
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("The Graph")
plt.legend()
plt.show()

#Save the graph
plt.savefig("output.jpg")
```
